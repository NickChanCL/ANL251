```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl


##First dataset: International Visitor Arrivals to Singapore
df = pd.read_csv("International Visitor Arrivals.csv",  delimiter = ',', skiprows=4, thousands=',')#reading the data while skipping the first 4 rows(first 4 rows have no data)

#generate boxplot to find outliers
col = list(df.columns)#selecting data needed
newdf = df[[col[0]] + col[-10:-1]]
newdf = newdf.iloc[1:62,]

newdf.boxplot()#generate boxplot

plt.title('Boxplot for Visitor Arrivals')#plot the title of the graph
plt.xticks(rotation=90)#rotate x-axis label

plt.show()

#generate line graph
data = df.iloc[0,-20:-1] #selecting the needed data from 2019 Jan to 2020 Jul to convert to float
pd.to_numeric(data, errors='coerce') #converting data from object(due to , in the data) to float

plt.title('International Visitor Arrivals')#plot the title of the graph

ax = data.plot()
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))#formatting the y axis to display by the thousands

plt.plot(12, df.iloc[0,-8], 'r*')#highlight key points
plt.annotate("23 Jan 2020 - First confirmed case of COVID-19 in SG", (12, df.iloc[0,-8]))#reason for highlighting key points

plt.plot(14, df.iloc[0,-6], 'r*')#highlight key points
plt.annotate("22 Mar 2020 - Prohibition against short-term visits to SG", (14, df.iloc[0,-6]))#reason for highlighting key points

plt.show()


##Second dataset: Hotel Revenue in Singapore
df = pd.read_csv("Hotel Statistics, Monthly.csv",  delimiter = ',', skiprows=4, thousands=',')#reading the data while skipping the first 4 rows(first 4 rows have no data)

data = df.iloc[0,1:7]#selecting data needed
data = data*1000 #converting data to dollars

ax = data.plot()
plt.title('Hotel Room Revenue')#plot the title of the graph
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))#setting y-axis to show in raw numbers

plt.plot(1, df.iloc[0,2]*1000, 'r*')#highlight key points
plt.annotate("22 Mar 2020 - Prohibition against short-term visits to SG", (1, df.iloc[0,2]*1000))#reason for highlighting key points

plt.show()


##Third dataset: Search frequency for Singapore flights
df = pd.read_csv("multiTimeline.csv",  delimiter = ',', skiprows=2)#reading the data while skipping the first 2 rows(first 2 rows have no data)
df.set_index('Week', inplace=True)#setting index to week

df.plot()

plt.title('Search frequency for Singapore flights')
plt.xticks(rotation=45)

plt.plot(19, df.iloc[19,0], 'r*')
plt.annotate("23 Jan 2020 - First confirmed case of COVID-19 in SG", (19, df.iloc[19,0]))

plt.plot(27, df.iloc[27,0], 'r*')
plt.annotate("22 Mar 2020 - Prohibition against short-term visits to SG", (27, df.iloc[27,0]))

plt.show()
```
